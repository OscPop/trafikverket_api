```python
import os
import requests
import json
from dotenv import load_dotenv

load_dotenv()


# Load API keys from .env file
SL_API_KEY = os.environ.get("SL_API_KEY")
TRAFIKVERKET_API_KEY = os.environ.get("TRAFIKVERKET_API_KEY")
```

```python
URL = f"https://api.sl.se/api2/trafficsituation.json?key={SL_API_KEY}"

resp = requests.get(URL).json()
resp
```

```python
resp["ResponseData"]["TrafficTypes"][1]
```

```python
for x in resp["ResponseData"]["TrafficTypes"]:
    print(x["Name"], x["Id"])
```

```python
# Ta ut alla events för röda linjen
for event in resp["ResponseData"]["TrafficTypes"][0]["Events"]:
    if event["TrafficLine"] == "Röda linjen":
        print(event["Message"])
```

```python
# Alla events för pendeltågslinjen
for event in resp["ResponseData"]["TrafficTypes"][1]["Events"]:
    if event["TrafficLine"] == "Pendeltåg":
        print(event["Message"])
```
